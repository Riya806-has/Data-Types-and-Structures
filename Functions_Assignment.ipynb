1. What is the difference between a function and a method in Python?
    *  In Python, functions and methods are both blocks of reusable code, but they have key differences based on how they are defined and used:


> Function


*   A function is a block of code defined using the def keyword (or lambda for anonymous functions).
*   It is not bound to any object.
*   Can be called independently and is typically defined at the module level.


```python
# Example:
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
```



> Method

* A method is a function that is associated with an object (usually an instance of a class).

* It is called using the dot (.) notation on an object.

* The first parameter is usually self, referring to the instance of the class.

```python
# Example:
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

g = Greeter()
print(g.greet("Bob"))
```

5. What are iterators in Python and how do they differ from iterables?

> What is an Iterable?

An iterable is any Python object that can be looped over (iterated through). It must implement the __iter__() method, which returns an iterator.

Examples of iterables:
* Lists, Tuples, Strings, Dictionaries, Sets

* Any object that implements __iter__() or __getitem__()

```python
# Example:

my_list = [1, 2, 3]
for item in my_list:
    print(item)
```



>  What is an Iterator?

An iterator is an object that represents a stream of data. It remembers its position and can return the next item with the next() function (or __next__() method). It is created by calling iter() on an iterable.


```python
# Example:

my_list = [1, 2, 3]
it = iter(my_list)

print(next(it))
print(next(it))
print(next(it))
```

6. Explain the concept of generators in Python and how they are defined.
    *  Generators in Python

  A generator in Python is a special type of function that produces a sequence of values over time, one at a time, using the yield keyword instead of return. It allows lazy evaluation, meaning it generates values on the fly and doesn't store them all in memory.

> Key Features of Generators

* Defined using def and yield

* Produces values one at a time (lazy evaluation)

* Automatically creates an iterator

* Saves memory and improves performance for large data

```python
# Defining a Generator Function
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
```

Calling this function:

```python
gen = count_up_to(3)
print(next(gen))
print(next(gen))
print(next(gen))
```

> How yield Works

* yield pauses the function, saving its state.

* On the next call (via next() or a loop), it resumes from where it left off.

```python
# Looping Through a Generator
for num in count_up_to(3):
    print(num)
```

7. What are the advantages of using generators over regular functions?
      *  Advantages of Using Generators Over Regular Functions in Python

Generators provide several performance and usability advantages, especially when working with large data sets or streams. Here's why they can be better than regular functions in certain scenarios:

> Memory Efficiency

* Generators don't store all values in memory—they produce one value at a time.

```python
# Regular Function (inefficient for large data):
def get_numbers():
    return [i for i in range(1_000_000)]
```

* Loads all 1 million numbers into memory at once.

```python
# Generator (efficient):

def generate_numbers():
    for i in range(1_000_000):
        yield i
```

* Produces one number at a time, saving memory.

> Lazy Evaluation

Generators compute values only when needed, which can significantly speed up programs when not all values are required.


```python
# Example:

def find_first_even(numbers):
    for num in numbers:
        if num % 2 == 0:
            return num

print(find_first_even(generate_numbers()))
```

* Stops as soon as the first even number is found—no need to process the rest.

>  Infinite Sequences

Generators can produce infinite streams of data safely.



```python
# Example:

def infinite_counter():
    i = 0
    while True:
        yield i
        i += 1
```

* Works forever (until stopped), unlike a list which would crash trying to store infinite values.

> Improved Performance

Because generators avoid overhead from memory allocation and data structure creation, they are faster for sequential data processing.

> Simpler Code for Iteration

Generators make it easy to write clean, readable code that yields sequences.

```python
# Example:

def read_lines(file_path):
    with open(file_path) as f:
        for line in f:
            yield line
```

* Efficient way to process a file line by line without loading it entirely into memory.

> Supports Pipelines (Chaining)

You can chain multiple generators together for step-by-step processing, similar to UNIX pipes.

```python
# Example:

gen = (x for x in range(10) if x % 2 == 0)
squared = (x*x for x in gen)
for num in squared:
    print(num)
```

8. What is a lambda function in Python and when is it typically used?

9. Explain the purpose and usage of the `map()` function in Python.
   *  The map() function in Python is used to apply a function to every item in an iterable (like a list or tuple) and return a new map object (which is an iterator).

> Purpose of map()

* To transform each element in an iterable using a given function.

* Avoids explicit loops (cleaner and often faster code).


> Syntax:

map(function, iterable)

* function: A function to apply (can be a regular or lambda function).

* iterable: A sequence (list, tuple, etc.) to process.

```python
# Example 1: Using map() with a built-in function

nums = [1, 2, 3, 4]
squared = map(pow, nums, [2]*len(nums))
print(list(squared))
```

```python
# Example 2: Using map() with a lambda function

nums = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, nums)
print(list(squared))
```

```python
# Example 3: Mapping Multiple Iterables

a = [1, 2, 3]
b = [4, 5, 6]
summed = map(lambda x, y: x + y, a, b)
print(list(summed))

```

> When to Use map()

* For simple, one-line transformations

* When using lambda or existing functions (like str.upper, int, round, etc.)

* To clean and transform data in functional pipelines

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
   *  Difference Between map(), filter(), and reduce() in Python

   These three functions — map(), filter(), and reduce() — are functional programming tools in Python. They work on iterables and apply functions in different ways.


>  1. map() – Transform each item

* Applies a function to every element in an iterable.

* Returns a new iterable with transformed items.

```python
# Example:
nums = [1, 2, 3, 4]
squared = map(lambda x: x**2, nums)
print(list(squared))
```

>  2. filter() – Keep only items that pass a test

* Applies a function that returns True/False to each item.

* Keeps only the items for which the function returns True.

```python
# Example:
nums = [1, 2, 3, 4]
evens = filter(lambda x: x % 2 == 0, nums)
print(list(evens))
```

> 3. reduce() – Combine all items into a single result

* Applies a function cumulatively to the items (from left to right), reducing the iterable to a single value.

* Available in the functools module.

```python
# Example:
from functools import reduce

nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product)
```

> When to Use:

* Use map() when you want to transform all items.

* Use filter() when you want to exclude some items.

* Use reduce() when you want to aggregate all items into one.

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b
# Example:
fibonacci = FibonacciIterator(10)
for num in fibonacci:
    print(num)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Optionally strip the newline character
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
# Example:
file_path = 'example.txt'
for line in read_file_line_by_line(file_path):
    print(line)
```

```python
# 10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.

# Function to remove vowels from a string
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_string))

# Example string
input_string = "Hello, World!"

# Removing vowels
result = remove_vowels(input_string)

print(result)
```
